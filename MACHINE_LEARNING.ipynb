```python
import pandas as pd

# List of file paths for the datasets
file_paths = [
    '1 - India_Economic_Growth_Historical_Data.csv', '2 -India_GDP_Growth_Rate_Historical_Data.csv', '3 - India_GDP_Historical_Data.csv', 
    '4 - India_GDP_Per_Capita_Historical_Data.csv', '5 - India_GNI_Historical_Data.csv', '6 - India_GNI_Per_Capita_Historical_Data.csv', 
    '7 - India_GNP_Historical_Data.csv', '8 - India_Government_Debt_as_percent_of_GDP_Historical_Data.csv', '9 - India_Inflation_Rate_Historical_Data.csv', 
    '10 - India_Manufacturing_Output_Historical_Data.csv'
]

# Column to be removed
column_to_remove = 'Unnamed: 0'

# Load all datasets into a list of DataFrames and remove the unwanted column
dataframes = [pd.read_csv(file_path).drop(columns=[column_to_remove], errors='ignore') for file_path in file_paths]

# Initialize the merged DataFrame with the first DataFrame in the list
merged_df = dataframes[0]

# Merge each DataFrame into the merged DataFrame iteratively with custom suffixes
for i, df in enumerate(dataframes[1:], start=2):
    suffix = f'_data{i}'
    merged_df = merged_df.merge(df, on='year', suffixes=('', suffix))

# Save the merged DataFrame to a new CSV file
#merged_df.to_csv('merged_dataset.csv', index=False)
```

```python

# Convert percentage columns to numerical values
for col in gdp.columns:
    if gdp[col].dtype == 'object' and '%' in gdp[col].iloc[0]:
        gdp[col] = gdp[col].str.rstrip('%').astype(float) / 100
```

```python
currency_columns = ['GDP', 'GNI', 'billions_of_us', 'GDP_Per_Capita', 'GNP', 'Per_Capita_GNP']  # Adjust this list based on your actual column names

# Apply string operations only to columns with object type
for col in currency_columns:
    if col in gdp.columns and gdp[col].dtype == 'object':
        gdp[col] = gdp[col].str.replace('[\$,B]', '', regex=True).replace('', '0').astype(float)
    elif col in gdp.columns:
        gdp[col] = gdp[col].astype(str).str.replace('[\$,B]', '', regex=True).replace('', '0').astype(float)
```
