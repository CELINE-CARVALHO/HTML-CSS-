# Categorizing items into food, bodycare and others

```python
import re
import pandas as pd

food = r'(?i)(?:drink|Food|food|chips|ice cream|biscuits|chocolates|Liquid|beverages|energy drink|nuts|namkeen|candy|coffee|tea)'
body_care = r'(?i)(?:Toletries|body care|Hair Care|Body care|skin care|Health|body spray)'

def categorizing(Category):
    if re.search(food, Category):
        return 'Food'
    elif re.search(body_care, Category):
        return 'Body Care'
    else:
        return 'Others'

df['Category'] = df['Category'].apply(categorizing)

# Print the DataFrame after categorizing
df
```

```python
df['Price'] = pd.to_numeric(df['Price'])
```
