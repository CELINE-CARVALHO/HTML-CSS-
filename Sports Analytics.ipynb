# <span style="color:darkorchid">**DISCRIPTION ABOUT THE DATASET**</span>

### this dataset provides comprehensive information about player performance during football matches, including offensive contributions (goals, assists, shots), defensive actions (clearances, fouls), and playing time, allowing for detailed analysis and insights into individual and team performance.

1. Name: This refers to the individual player's name.
2. Team Name: This indicates the team the player belongs to.
3. Playtime: This represents the total duration (in minutes) the player was on the field during the match.
4. Shots on target: This counts the number of times a player attempted to score a goal, and the ball either hit the target (goalkeeper saved it) or went inside the net.
5. Fouls: This indicates the number of fouls a player committed during the game. Fouls are penalized for breaking the rules, like tripping or pushing an opponent.
6. Total no. of shots: This is the total number of times a player attempted to kick the ball towards the goal, regardless of whether it was on target or not.
7. Goals scored: This simply counts the number of goals a player successfully scored during the match.
8. Clearance: This refers to the number of times a player intervened to prevent the opposing team from getting control of the ball, often by heading the ball out of their defensive zone.
9. Assist: An assist is credited to a player who makes a pass that directly leads to a teammate scoring a goal.
10. Course: which course fron the department does he belong(bsc ds ,bsc ea, msc ea and ds)

### <span style="color:darkred">***analysis***</span>

```python
players_with_shots = data[data["Shorts on target"] > 0]

# Get player names and goals scored
players_and_goals = players_with_shots[["Name", "Goals scored"]]

# Print the results
print(players_and_goals.to_string(index=False))
```

### In this i have analysed on the player that tried to shot on the target but some scored the goal and some did not score

```python
def shot_efficiency(shots_on_target, total_shots):
  if total_shots == 0:
    return 0  # Avoid division by zero
  return shots_on_target / total_shots

# Add a 'Shot Efficiency' column
data['Shot Efficiency'] = data.apply(lambda row: shot_efficiency(row['Shorts on target'], row['Total no.of shots']), axis=1)

# Top Scorer
top_scorer = data.loc[data['Goals scored'].idxmax()]

# Best Shot Efficiency
best_efficiency = data.loc[data['Shot Efficiency'].idxmax()]

# Top Offensive Contributor (combined goals and assists)
offensive_contribution = data['Goals scored'] + data['Assist']
top_offensive = data.loc[offensive_contribution.idxmax()]

# Top Defensive Contributor (most clearances)
top_defensive = data.loc[data['Clearance '].idxmax()]

# Print results
print("Top Scorer:")
print(top_scorer[['Name', 'Goals scored']].to_string(index=False))

print("\nBest Shot Efficiency:")
print(best_efficiency[['Name', 'Shot Efficiency']].to_string(index=False))

print("\nTop Offensive Contributor:")
print(top_offensive[['Name', 'Goals scored', 'Assist']].to_string(index=False))

print("\nTop Defensive Contributor:")
print(top_defensive[['Name', 'Clearance ']].to_string(index=False))
```

This information can be used to evaluate player performance and identify key contributors to team success.
